```python
jd = pd.merge(ol,od, how='left', on=['order_id', 'order_id'])
jd.head()
```

```python
# ngr = jd.groupby('country').count()
# ngr
```

```python
ngr = jd.groupby('country').agg(rata_rata = ('sales','mean'),
                               total = ('sales','sum'))
ngr
```

```python
prd = pd.pivot_table(jd, index =['category'],
                    values=['sales'],
                    aggfunc={'sales':['min','max']})
prd
```

```python
prd = jd.groupby('category').agg(penjualan_min = ('sales','min'),
                               penjualan_max = ('sales','max'))
prd
```

```python
prc = jd.groupby('customer_name').agg(total_kontribusi = ('Total_profit','sum'))
prc.head()
```
